```python
def tenure_group(tenure):
  if tenure <= 12:
      return 'Novo Cliente (1-12 meses)'
  elif tenure <= 36:
      return 'Cliente MÃ©dio (13-36 meses)'
  elif tenure <= 60:
      return 'Cliente Longo (37-60 meses)'
  else:
      return 'Cliente Fiel (61+ meses)'

dados['customer.tenure2'] = dados['customer.tenure'].apply(tenure_group)

fig, ax = plt.subplots(figsize=(10, 5))
# Usamos multiple='dodge' para colocar as barras lado a lado
ax = sns.histplot(data=dados, x='customer.tenure2', hue='Churn', multiple='dodge',
                  shrink=.8)

# Adiciona e formata os tÃ­tulos e labels
ax.set_title('Contagem de EvasÃ£o por Tempo de Contrato', fontsize=16, pad=20)
ax.set_xlabel('Tempo de Contrato', fontsize=12)
ax.set_ylabel('Quantidade de Clientes', fontsize=12)

# Adiciona os rÃ³tulos de dados em cima de cada barra
for container in ax.containers:
    ax.bar_label(container, padding=3, fontsize=10)

# Melhora a legenda
plt.legend(title='EvasÃ£o', labels=['NÃ£o', 'Sim'])
plt.grid(linestyle=':')
sns.despine()
plt.tight_layout()
plt.show()
```

### **RelatÃ³rio de AnÃ¡lise ExploratÃ³ria â€“ Fatores de EvasÃ£o de Clientes na TelecomX**


#### **1. IntroduÃ§Ã£o: O Problema de NegÃ³cio**

A evasÃ£o de clientes, ou "Churn", Ã© uma mÃ©trica crÃ­tica para empresas de serviÃ§os recorrentes, como as do setor de telecomunicaÃ§Ãµes. Perder um cliente nÃ£o representa apenas a perda de receita futura, mas tambÃ©m implica em custos mais elevados para adquirir novos clientes em comparaÃ§Ã£o com os custos de retenÃ§Ã£o dos existentes.

O presente estudo foi realizado com o objetivo de analisar a base de dados da TelecomX para identificar os principais fatores e padrÃµes de comportamento que estÃ£o associados Ã  decisÃ£o de um cliente cancelar seus serviÃ§os. A compreensÃ£o desses fatores Ã© o primeiro passo para desenvolver estratÃ©gias de retenÃ§Ã£o mais eficazes, personalizar ofertas e, consequentemente, aumentar a lucratividade e a sustentabilidade do negÃ³cio.

#### **2. Processo de AnÃ¡lise dos Dados**

A anÃ¡lise seguiu um processo estruturado, utilizando o notebook `telecomx_br.py`, que pode ser dividido em trÃªs fases principais:

1.  **ExtraÃ§Ã£o e PreparaÃ§Ã£o dos Dados:**
    * Os dados foram extraÃ­dos de um arquivo JSON.
    * A estrutura aninhada do JSON foi normalizada para um formato de tabela (DataFrame) para facilitar a anÃ¡lise.
    * Foi realizada uma limpeza nos dados: a coluna de cobranÃ§a total (`Charges.Total`) foi convertida de texto para um formato numÃ©rico, e valores inconsistentes (espaÃ§os em branco) foram tratados como nulos.
    * Registros que nÃ£o continham informaÃ§Ã£o sobre a evasÃ£o (`Churn`) foram removidos para garantir a precisÃ£o da anÃ¡lise.

2.  **Engenharia de Atributos:**
    * Para aprofundar a anÃ¡lise, novas colunas (features) foram criadas a partir dos dados existentes:
        * `account.Charges.Daily`: Calculado dividindo o custo mensal por 30, para obter uma visÃ£o do custo diÃ¡rio do serviÃ§o.
        * `NumAddonServices`: Uma contagem do nÃºmero total de serviÃ§os adicionais contratados por cada cliente (ex: seguranÃ§a online, backup, suporte tÃ©cnico, etc.).
        * `customer.tenure2` e `account.Charges.Daily2`: CategorizaÃ§Ã£o do tempo de contrato e do custo diÃ¡rio em grupos, facilitando a visualizaÃ§Ã£o e interpretaÃ§Ã£o dos grÃ¡ficos.

3.  **AnÃ¡lise ExploratÃ³ria Visual:**
    * Foram gerados diversos grÃ¡ficos para visualizar a relaÃ§Ã£o entre as variÃ¡veis do cliente (tipo de contrato, tempo de serviÃ§o, custo, etc.) e a taxa de evasÃ£o. Essa abordagem visual permitiu a rÃ¡pida identificaÃ§Ã£o de tendÃªncias e padrÃµes significativos.

#### **3. Resultados e Descobertas da AnÃ¡lise**

A anÃ¡lise exploratÃ³ria revelou insights claros sobre o perfil dos clientes que tendem a deixar a empresa.

* **Perfil Geral da EvasÃ£o:**
    * Da base de clientes analisada, **1.869 clientes evadiram**, enquanto **5.174 permaneceram**. Isso representa uma taxa de churn de aproximadamente 26,5% na amostra, um nÃºmero significativo que justifica a investigaÃ§Ã£o.

* **Tipo de Contrato Ã© o Fator Mais CrÃ­tico:**
    * A anÃ¡lise demonstrou que o tipo de contrato Ã© o indicador mais forte de evasÃ£o. Clientes com **contratos mensais** apresentam uma taxa de churn drasticamente maior em comparaÃ§Ã£o com aqueles com contratos de longo prazo (anuais ou bianuais). A flexibilidade do contrato mensal facilita o cancelamento do serviÃ§o.

* **Tempo de Contrato (Tenure) e a Lealdade do Cliente:**
    * Existe uma relaÃ§Ã£o direta entre o tempo de permanÃªncia do cliente e sua probabilidade de evasÃ£o. **Clientes mais novos (com atÃ© 12 meses de contrato) sÃ£o muito mais propensos a cancelar o serviÃ§o**. A taxa de churn diminui consideravelmente Ã  medida que o cliente se torna mais antigo e fiel.

* **ServiÃ§os de Internet e Custo DiÃ¡rio:**
    * Clientes que assinam o serviÃ§o de internet via **Fibra Ã“ptica** tÃªm uma taxa de evasÃ£o maior do que os assinantes de DSL. Isso pode ser um indicativo de problemas de concorrÃªncia, preÃ§o ou qualidade percebida nesse serviÃ§o especÃ­fico.
    * A anÃ¡lise de custo diÃ¡rio mostra que a evasÃ£o Ã© mais acentuada em faixas de custo mais altas, especialmente quando combinada com contratos mensais.

* **InfluÃªncia dos ServiÃ§os Adicionais:**
    * Clientes com um **maior nÃºmero de serviÃ§os adicionais contratados tendem a evadir menos**. Clientes que nÃ£o evadiram possuem, em mÃ©dia, mais serviÃ§os adicionais do que aqueles que evadiram. Isso sugere que quanto mais integrado o cliente estÃ¡ ao ecossistema de serviÃ§os da empresa, maior Ã© a sua lealdade.

* **Fatores DemogrÃ¡ficos com Menor Impacto:**
    * **GÃªnero:** NÃ£o foi identificada uma diferenÃ§a significativa na taxa de evasÃ£o entre clientes do sexo masculino e feminino.
    * **Idade (SeniorCitizen):** Embora clientes idosos (65 anos ou mais) representem uma parcela menor da base total, eles possuem uma taxa de evasÃ£o proporcionalmente maior do que os clientes mais jovens.

#### **4. ConclusÃ£o e RecomendaÃ§Ãµes EstratÃ©gicas**

A anÃ¡lise dos dados da TelecomX permite traÃ§ar um perfil claro do cliente com alto risco de evasÃ£o: **um cliente recente (baixo tempo de contrato), com contrato do tipo mensal, que possui poucos ou nenhum serviÃ§o adicional contratado e, possivelmente, um assinante do serviÃ§o de Fibra Ã“ptica.**

Com base nessas descobertas, as seguintes aÃ§Ãµes estratÃ©gicas sÃ£o recomendadas:

1.  **Incentivar Contratos de Longo Prazo:** Criar campanhas ativas e oferecer descontos ou benefÃ­cios claros para que clientes de contratos mensais migrem para planos anuais ou bianuais.
2.  **Foco na RetenÃ§Ã£o de Novos Clientes:** Desenvolver um programa de "Boas-vindas" e acompanhamento prÃ³ximo para clientes nos primeiros 12 meses, perÃ­odo em que o risco de churn Ã© maior. Ofertas personalizadas e um suporte tÃ©cnico proativo podem fazer a diferenÃ§a.
3.  **Promover Pacotes de ServiÃ§os (Bundles):** Incentivar a contrataÃ§Ã£o de serviÃ§os adicionais, demonstrando o valor agregado e oferecendo descontos na aquisiÃ§Ã£o de pacotes. Isso aumenta a "pegada" do cliente na empresa, tornando a saÃ­da mais difÃ­cil.
4.  **Investigar a EvasÃ£o na Fibra Ã“ptica:** Realizar uma anÃ¡lise mais aprofundada sobre os clientes de Fibra Ã“ptica para entender as causas da alta evasÃ£o. Pode envolver pesquisa de satisfaÃ§Ã£o, anÃ¡lise da concorrÃªncia e revisÃ£o da estrutura de preÃ§os e qualidade do serviÃ§o.

Esta anÃ¡lise exploratÃ³ria fornece uma base sÃ³lida para a tomada de decisÃµes orientada por dados, permitindo que a TelecomX direcione seus esforÃ§os de retenÃ§Ã£o de forma mais eficiente e assertiva.
